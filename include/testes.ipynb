```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
delay = 20
exp = pd.read_csv("exp02/pwm_140_15:00:05.csv")
plt.plot(exp['Time'] + delay*5, exp['Left'])
plt.plot(exp['Time'], exp['Right'])
# plt.plot(exp['Time'] + delay, exp['Left'].diff().rolling(6).mean())
# plt.plot(exp['Time'], exp['Right'].diff().rolling(6).mean())
plt.legend(['Motor esquerdo', 'Motor Direito'])
plt.ylabel('Velocidade')
plt.xlabel('Tempo (ms)')
plt.gca().twinx()
plt.plot(exp['Time'], exp['Left'].shift(delay) - exp['Right'], 'r--')
plt.ylabel('Ã‚ngulo')
plt.title(f'Atraso no motor esquerdo ({delay*5} ms)')
```
