```python
import numpy as np
import pandas as pd


v = 5*120/(32*np.pi)
translational = np.ones(1000)*v
rotational = np.zeros_like(translational)

rotational[np.r_[
    150:200,
    300:350,
    650:700,
    800:850
]] = 120*115/(2*32*50)

pd.Series(rotational).rolling(40).mean().plot()
pd.Series(translational).plot()
```

```python
import matplotlib.pyplot as plt

x = y = np.zeros_like(translational)

theta = np.cumsum(rotational)*32*np.pi/(120*115)

for i in np.arange(1, 1000):
    x[i] = x[i-1] + np.cos(theta[i])*v
    y[i] = y[i-1] + np.sin(theta[i])*v

plt.plot(x, y)
```
