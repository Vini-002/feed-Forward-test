```python
# exp = data.loc[946:10851, :]
```

```python
exp['y'] = 0
exp['x'] = 0
exp['theta'] = (exp['Left'] - exp['Right'])*32*np.pi/(120*115)
```

```python
# exp = exp.drop(exp.columns[0], axis='columns')
# exp['Left'] = exp['Left']*32*np.pi/120
# exp['Right'] = exp['Right']*32*np.pi/120

exp['DeltaLeft'] = exp['Left'].rolling(20).mean().diff()   # *32*np.pi/120
exp['DeltaRight'] = exp['Right'].rolling(20).mean().diff() # *32*np.pi/120
exp = exp.dropna()
# exp
```

```python
x = [0]*25001
y = [0]*25001
for i, (l, r, t) in enumerate(zip(exp['Left'].to_list(), exp['Right'].to_list(), exp['theta'].to_list())):
    x[i + 1] = x[i] + ((l + r)/2)*np.cos(t)
    y[i + 1] = y[i] + ((l + r)/2)*np.sin(t)
```

```python
import matplotlib.pyplot as plt

plt.plot(x, y)
ax = plt.gca()
plt.gca().invert_xaxis()
ax.set_aspect('equal')
# ax.set_ylim((-10000, 10000))
```
